```python
import requests
from bs4 import BeautifulSoup


response = requests.get("https://ru.wikipedia.org/wiki/_")
print(response.status_code)

soup = BeautifulSoup(response.content, "html.parser")
a_title = soup.find_all("a", class_="tm-header__logo tm-header__logo_en")
for element in a_title:
    print(element.text)
```
